```python

from sklearn.svm  import LinearSVC
from sklearn.linear_model import LogisticRegression
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
cm_bright = ListedColormap(['#FF0000', '#0000FF'])
cm = plt.cm.RdBu

# Function to plot decision boundry, restricted to 2-d plots ie 2 features
def plot_decision_boundry(model,X):
    x1, x2 = np.meshgrid(np.arange(X[:,0].min() -.5, X[:,0].max() +.5 , 0.05),
                         np.arange(X[:,1].min() -.5, X[:,1].max() +.5,  0.05))
    y_plot_pred_probs = model.decision_function(np.c_[x1.ravel(), x2.ravel()])#[:,1]

    plt.contourf(x1,
                 x2,
                 y_plot_pred_probs.reshape(x1.shape), cmap=cm, alpha=.7)
    plt.xlabel('feature 1')
    plt.ylabel('feature 2')
    #plt.show()
    
    #y_scores = model.decision_function(np.c_[x1.ravel(), x2.ravel()])
    #y_prob2 = np.exp(y_scores) / (np.exp(y_scores) +1) # which is nothing but probability of class 1

    #print(y_scores.max())
    #plt.scatter(y_scores, y_prob2, c="g")
    #plt.xlabel('Scores (log-odds)')
    #plt.ylabel('Probability')
    #plt.show()
```

# SVM vs Logistic Regression

```python
# Lets create a manual dataset
np.random.seed(10)
n = 10
X = np.concatenate((2*np.random.rand(n,2)+1,3*np.random.rand(n,2)+5))
y = np.concatenate((np.repeat(0,n),np.repeat(1,n)))
```

## Logistic Regression vs linear SVM
(Without Regualrization)

As you can see when both of the classification models (Logistic and SVM) has no regularization 
* The max margin better decision boundry is learned by SVM and not Logistic 
* Weights are more for logistic regression for similar parameters

Because the formulation of SVM is based on maximizing the geometrical margin of the decision boundry while in case of logistic it is maximizing the likelihood, which tries to overfit.

```python
# Logistic Regression
log_clf = LogisticRegression(C=100000000) 
log_clf.fit(X,y)
print(log_clf.coef_, log_clf.intercept_) 

#y_train_proba = log_reg.predict_proba(X)
plt.scatter(X[:,0],X[:,1],c=y,cmap=cm_bright,
            edgecolor='k')

plot_decision_boundry(log_clf,X)
plt.show()

# Linear Kernel SVM
linSVM_clf = LinearSVC(C=100000000,loss='hinge', max_iter=10000)
linSVM_clf.fit(X,y)
print(linSVM_clf.coef_, linSVM_clf.intercept_) 

#y_train_proba = log_reg.predict_proba(X)
plt.scatter(X[:,0],X[:,1],c=y,cmap=cm_bright,
            edgecolor='k')

plot_decision_boundry(linSVM_clf,X)
plt.show()

```

# SVC vs LinearSVC

We can also use SVC to create a linear SVM model. SVC also provides us with support vectors in the output. Creating a model for above data and dropping support vectors too.


You should always try linear kernel first, remmber that LinearSVC is much faster than SVC(kernel='Linear)

```python

from sklearn.svm import SVC
# Linear Kernel SVM
linSVM_clf = SVC(kernel='linear',  C=10000000, probability=True)
linSVM_clf.fit(X,y)

plt.scatter(X[:,0],X[:,1],c=y,cmap=cm_bright,
            edgecolor='k')
plot_decision_boundry(linSVM_clf,X)

sup_vec = linSVM_clf.support_vectors_
sup_vec_label = y[linSVM_clf.support_]
#sup_vec = poly_svc_clf['Scaler'].inverse_transform(sup_vec) # Inversing the standardization we did at the time of training
plt.scatter(sup_vec[:,0],sup_vec[:,1],c = sup_vec_label, cmap=plt.cm.Set2,
            edgecolor='k')
plt.show()
```

## How outlier affects the SVM decision boundry

```python
from sklearn.svm import SVC
# Now add an outlier to the above data, class 1 instance very near to class 0.
X2 = np.concatenate((X, np.array([[3.1,3.1]])))
y2 = np.concatenate((y,[1]))

#linSVM_clf2 = LinearSVC(C=100000000,loss='hinge', max_iter=1000000)
linSVM_clf_out = SVC(kernel='linear',  C=1000000000, probability=True)
linSVM_clf_out.fit(X2,y2)

plt.scatter(X2[:,0], X2[:,1], c=y2, cmap=cm_bright, edgecolor='k')
plot_decision_boundry(linSVM_clf_out,X2)

sup_vec = linSVM_clf_out.support_vectors_
sup_vec_label = y2[linSVM_clf_out.support_]
#sup_vec = poly_svc_clf['Scaler'].inverse_transform(sup_vec) # Inversing the standardization we did at the time of training
plt.scatter(sup_vec[:,0],sup_vec[:,1],c = sup_vec_label, cmap=plt.cm.Set2, edgecolor='k')
plt.show()
```

# Non-Linear classification problem

```python
# Lets make a polynomial classifier for non-linear classification problem

from sklearn.datasets import make_moons
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.metrics import accuracy_score

# Generating the data
# Training Set
X,y = make_moons(n_samples = 1000, noise =0.2)
# Test Set
X_test,y_test = make_moons(n_samples = 1000, noise =.2)

plt.scatter(X[:,0],X[:,1],c=y,cmap=cm_bright,
            edgecolor='k')
plt.show()
```

## Using polynomial kernal

```python
from sklearn.svm import SVC
from matplotlib.colors import ListedColormap
cm_bright = ListedColormap(['#FF0000', '#0000FF'])
# Training the model
poly_svc_clf = Pipeline([
    ("Scaler" , StandardScaler()),
    ("svc_clf" , SVC(kernel='poly', degree=3, coef0=1, C=1, probability=False)) 
    # Hyperparameters : 
     ## coef0: is necessary, it means how much the model is influenced by higher degree than the lower degree features
     ## C: Regularization parameter, it is inversely proprtional to the strength of regularization,
           #penalty square l2 pernalty, less the value more will be the margin voilations and bigger will be the margin
])                                             
poly_svc_clf.fit(X,y)

y_pred = poly_svc_clf.predict(X)            # Train data
y_test_pred = poly_svc_clf.predict(X_test)  # Test data

print('--> Accuracy of the model on Trainset: ',accuracy_score(y,y_pred))
print('--> Accuracy of the model on Trainset: ',accuracy_score(y_test,y_test_pred))

plot_decision_boundry(poly_svc_clf,X)

# Plotting the training data
plt.scatter(X[:,0],X[:,1],c=y,cmap=cm_bright,
            edgecolor='k',alpha=.2)

# Plotting the support vectors
sup_vec = poly_svc_clf['svc_clf'].support_vectors_
sup_vec_label = y[poly_svc_clf['svc_clf'].support_]
sup_vec = poly_svc_clf['Scaler'].inverse_transform(sup_vec) # Inversing the standardization we did at the time of training

plt.scatter(sup_vec[:,0],sup_vec[:,1],c = sup_vec_label, cmap=plt.cm.Set2,
            edgecolor='k')
plt.show()
```

## Using RBF Kernel

```python
from sklearn.svm import SVC
from matplotlib.colors import ListedColormap
cm_bright = ListedColormap(['#FF0000', '#0000FF'])
# Training the model
poly_svc_clf = Pipeline([
    ("Scaler" , StandardScaler()),
    ("svc_clf" , SVC(kernel='rbf', gamma = 2, C=10, probability=False)) 
    # Hyperparameters : 
     ## gamma: Makes the bell shape curve of a landmark narrower, means each isnstance's range of influence is smaller,
              # the decison boundry ends up being more irregular, wiggling around individual instances.
              # so overall it acts like a regularization parameter like C.
     ## C: Regularization parameter, it is inversely proprtional to the strength of regularization,
           #penalty square l2 pernalty, less the value more will be the margin voilations and bigger will be the margin
])                                             
poly_svc_clf.fit(X,y)

y_pred = poly_svc_clf.predict(X)            # Train data
y_test_pred = poly_svc_clf.predict(X_test)  # Test data

print('--> Accuracy of the model on Trainset: ',accuracy_score(y,y_pred))
print('--> Accuracy of the model on Trainset: ',accuracy_score(y_test,y_test_pred))

plot_decision_boundry(poly_svc_clf,X)

# Plotting the training data
plt.scatter(X[:,0],X[:,1],c=y,cmap=cm_bright,
            edgecolor='k',alpha=.2)

# Plotting the support vectors
sup_vec = poly_svc_clf['svc_clf'].support_vectors_
sup_vec_label = y[poly_svc_clf['svc_clf'].support_]
sup_vec = poly_svc_clf['Scaler'].inverse_transform(sup_vec) # Inversing the standardization we did at the time of training

plt.scatter(sup_vec[:,0],sup_vec[:,1],c = sup_vec_label, cmap=plt.cm.Set2,
            edgecolor='k')
plt.show()
```

## Which kernel to use?

With so many kernel to choose from ie linear, poly, rbf, string etc, how can use decide which one to use?

As a rule of thumb : 

* You should always try linear kernel first, rember that LinearSVC is much faster than SVC(kernel='Linear), especially if the training set is very large or if it has plenty of features.
* If the training set is not too large, you should try Gaussian RBF kernel, it works well in most cases.
* If have spare time, you can try other kernels using cross-validation and grid search.
