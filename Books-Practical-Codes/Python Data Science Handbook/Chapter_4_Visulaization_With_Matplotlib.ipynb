## General Matplotlib Tips

```python
import numpy as np
```

```python
# Importing
#import matplotlib as mpl
import matplotlib.pyplot as plt

#%matplotlib inline
# Setting Styles
## We will use the plt.style directive to choose appropriate aesthetic styles for our figures. 
## Here we will set the classic style, which ensures that the plots we create use the classic Matplotlib style:
plt.style.use('classic')

# how() or No show()? How to Display Your Plots
## plotting from script
### plt.show() starts an event loop, looks for all currently active figure objects, and opens one or more 
### interactive windows that display your figure or figures.
x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x),'--')

plt.show()

## Point to Note : One thing to be aware of: the plt.show() command should be used only once per Python session,
## and is most often seen at the very end of the script. Multiple show() commands can lead to unpredictable 
## backend-dependent behavior, and should mostly be avoided.


# Saving figures to file

fig = plt.figure()
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

fig.savefig('temp.png')
```

## Simple plot line

```python
# set another style for plots
plt.style.use('seaborn-whitegrid')


# In Matplotlib, the figure (an instance of the class plt.Figure) can be thought of as a single container that 
# contains all the objects representing axes, graphics, text, and labels. 
# The axes (an instance of the class plt.Axes) is what we see above: a bounding box with ticks and labels, which
# will eventually contain the plot elements that make up our visualization.
fig = plt.figure()
ax = plt.axis()

# Once the axis are setup as above we can use ax.plot instead of plt.plot to plot
x = np.linspace(0, 10, 1000)
plt.plot(x, np.sin(x))
```

```python
# alternatively, you dont have to explictely instatiate figure or axis obljects, pyplot can directly do it
plt.plot(x, np.sin(x))
```

```python
### Colors

plt.plot(x, np.sin(x - 0), color='blue')        # specify color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#F000F0')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

```python
### linestyle

plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle='-.');  # dotted
```

```python
# the linestyle and color codes can be combined into a single non-keyword argument to the plt.plot() function:


# For short, you can use the following codes:
plt.plot(x, x + 4, '-g')  
plt.plot(x, x + 5, '--r') 
plt.plot(x, x + 6, '-.b') 
plt.plot(x, x + 7, ',c') 
```

#### Adjusting the Plot: Axes Limits

Matplotlib does a decent job of choosing default axes limits for your plot, but sometimes it's nice to have finer control. The most basic way to adjust axis limits is to use the plt.xlim() and plt.ylim() methods:

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')

# axis limits
plt.xlim(-1, 11)
plt.ylim(1.5, -1.5)   # reverse the order of interval to mirror the graph
# plt.axis([-1, 11, -1.5, 1.5]);  # also can be set with plt.axis method, other arguments like 'tight','equal' can be given

# labels
plt.title('Temporary Title- Sign Graph')
plt.xlabel("x")
plt.ylabel("sin(x)");

# set legent
plt.legend()
```

In the object-oriented interface to plotting, rather than calling these functions individually, it is often more convenient to use the ax.set() method to set all these properties at once:

```python
ax = plt.axes()
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
ax.set(xlim=(0, 10), ylim=(-2, 2),
       xlabel='x', ylabel='sin(x)',
       title='A Simple Plot');
```

## Scatter plot

scatter plots can be plotted using plt.plot, define a point marker, or plt.scatter.

plt.scatter also gives power to vary size and color for every individual point in plot.

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='black');
```

```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's','p', 'd','--','-ok']:#'-ok' represent line, point and color
    plt.plot(rng.rand(5), rng.rand(5), marker,
             label="marker='{0}'".format(marker))
plt.legend(numpoints=1)
plt.xlim(0, 1.8);
```

```python

# Extra merker options
plt.plot(x, y,  marker='-p',color='gray',
         markersize=15, linewidth=4,
         markerfacecolor='white',
         markeredgecolor='gray',
         markeredgewidth=2)
plt.ylim(-1.2, 1.2);
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5,
            cmap='viridis')
plt.colorbar();  # show color scale
```
